```python
def lxx(x):
    mean = sum(x) / len(x) 
    tar = sum((xi - mean) ** 2 for xi in x) 
    return tar

def lxy(x,y):
    mean_x=sum(x)/len(x)
    mean_y=sum(y)/len(y)
    i=0
    tar=0
    while i<len(x):
        tar+=(x[i]-mean_x)*(y[i]-mean_y)
        i+=1
    return tar

def ST(x):
    mean = sum(x) / len(x) 
    tar = sum((xi - mean) ** 2 for xi in x) 
    return tar

def SE(y,y_hat):
    i=0
    tar=0
    while i<len(y):
      tar+=(y[i]-y_hat[i])**2
      i+=1
    return tar

def SR(y,y_hat):
    mean_y=sum(y)/len(y)
    i=0
    tar=0
    while i<len(y_hat):
        tar+=(y_hat[i]-mean_y)**2
        i+=1
    return tar
```

```python
from matplotlib import pyplot as plt
import numpy as np
x=[820,930,1050,1300,1440,1500,1600,1800,2000,2700,3000,4000]
y=[750,850,920,1050,1200,1200,1300,1450,1560,2000,2000,2200]
fig,ax=plt.subplots(figsize=(8,8))
ax.set_xlabel("Family Average Income")
ax.set_ylabel("Average Food Expenditure")
ax.set_title("FAI vs AFE")
ax.scatter(x,y,color="red",marker="+")
plt.savefig("Q4.png")
```

```python
print("lxx:",lxx(x))
print("lxy:",lxy(x,y))
mean_x=sum(x)/len(x)
mean_y=sum(y)/len(y)
print("mean_x:",sum(x)/len(x))
print("mean_y:",sum(y)/len(y))
beta1_hat=lxy(x,y)/lxx(x)
beta0_hat=mean_y-beta1_hat*mean_x
print("beta0:",beta0_hat)
print("beta1:",beta1_hat)
y_hat=beta0_hat+beta1_hat*np.array(x)
print("ST:",ST(y))
print("SR:",SR(y,y_hat))
print("SE:",SE(y,y_hat))
```
